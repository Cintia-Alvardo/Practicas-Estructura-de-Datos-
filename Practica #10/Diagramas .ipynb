```python
#Tabla
import pandas as pd
import matplotlib as plt

fbk = ['Facebook', 2449, 2006]
twt = ['Twitter', 339, 2006]
ig = ['Instragram', 1000, 2000]
yt = ['YouTube', 2000, 2005]
wsp = ['Whatsapp', 1600, 2009]

lista_redes = [fbk, twt, ig, yt, wsp]
df_redes = pd.DataFrame(lista_redes, columns = ['Nombre', 'Cantidad', 'Año'])
df_redes
```

```python
plt.plot(df_redes['Nombre'], df_redes['Cantidad'])
plt.title('Diagrama de Lineas')
plt.xlabel('Redes')
plt.ylabel('Cantidad')
plt.show()
```

```python
df_redes_ordenadas = df_redes.sort_values('Cantidad', ascending = False)
plt.bar(df_redes_ordenadas['Nombre'], df_redes_ordenadas['Cantidad'], color = ['c', 'm', 'k'])
plt.title('Diagrama de Lineas')
plt.xlabel('Redes')
plt.ylabel('Cantidad')
plt.show()
```
